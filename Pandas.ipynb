```python
#1.apply()
def last_four(num):
    return str(num)[-4:]
```

```python
new_data['last_four']=new_data['CC Number'].apply(last_four)
```

```python
new_data.head()
```

```python
def yelp(price):
    if price<10:
        return '$'
    elif price>=10 or price<30:
        return '$$'
    else:
        return '$$$'
```

```python
new_data['Price Range']=new_data['total_bill'].apply(yelp)
```

```python
new_data.head()
```

```python
(new_data["Price Range"]=='$$$') is True
```

```python
#Using apply() with multiple columns
# i.e use np.vectorize()

def tip_genorisity(total_bill,tip):
    if tip/total_bill> 0.25:
        return "Generous"
    
    else:
        return "Other"
```

```python
#Now broadcasting the function to multiple columns
import numpy as np
new_data['Tip Generosity']=np.vectorize(tip_genorisity)(new_data['total_bill'],new_data['tip'])
```

```python
new_data
```

```python
# 2. map(): Used to substitute a value in a series with another value
dict_new={'Male':'M',"Female":"F"}
```
