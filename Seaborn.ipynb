## 2. Distribution Plots
A distribution plot is used to see how a feature is distributed

There are 3 types of Distribution plots:
1. Rug plots: It marks a data point on the x-axis
2. Histogram: It tells us how many data points are there in the given          range (bins).
3. KDE: It is used to estimate the PDF of a random variable

```python
#1. Rug plot
sns.rugplot(x="salary",data=df,height=0.5)
```

```python
#2. displot or histplot
sns.displot(x='salary', data=df,bins=10,rug=True,kde=True,color="red")
```

```python
#KDE plot
sns.set(style="ticks")
sns.kdeplot(x='salary',data=df,clip=[4000,160000],bw_adjust=1,shade=True)
```

## Categorical Plots

* Categorical plots are used to find out statistical information per category.
* They are visual representations of the groupby().

Categorical Plots-Stat Estimation
1. Count Plot
2. Bar Plot

Categorical Plots-Plot Distribution
1. Box Plot
2. Violin Plot
3. Swarm Plot
4. Boxen Plot

```python
# Catergorical Plots- Stat Estimation
#1. Count Plot 
sns.set(style="darkgrid")
plt.figure(figsize=(10,4),dpi=400)
sns.countplot(x='division',data=df,hue="level of education")
```

```python
#2. Bar plot
plt.figure(figsize=(10,4))
sns.barplot(x='division',y="salary",data=df,estimator=np.mean,ci='sd',hue='level of education')
plt.legend(bbox_to_anchor=(1.05,0.5))
```

```python
# Catergorical Plots- Distributions
# 1. Boxplot: Represents IQR,Median and whiskers and outliers
df1=pd.read_csv(r"C:\Users\Vadiraj R Adabaddi\Desktop\Data Science\ml\05-Seaborn\StudentsPerformance.csv")
```

```python
df1
```

```python
plt.figure(figsize=(10,6),dpi=200)
sns.boxplot(x="math score",y="parental level of education" ,data=df1,hue='gender')
```

```python
# 2. Violin plot: It is essentially a combination of Bar plot and KDE plot
plt.figure(figsize=(10,8),dpi=200)
sns.violinplot(x="math score",y="parental level of education" ,data=df1,hue='gender',split=True,inner='quartile') #inner=stick and box
```

```python
# 3. Swarm plot: A strip plot which is adjusted to avoid overlapping of points
# Note: A strip plot is a sctter plot with categories
plt.figure(figsize=(10,8),dpi=200)
sns.swarmplot(x="math score",y="parental level of education" ,hue='gender', dodge=True,data=df1,size=2)
```

```python
# 4. Boxen plot: Its similar to box plot. But quantiles are used instead of quartiles. It is rarely used.
plt.figure(figsize=(10,8),dpi=200)

sns.boxenplot(x="math score",y="parental level of education" ,data=df1)
```
