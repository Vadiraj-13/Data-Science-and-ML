# Matplotlib
* Known as "Grandfather library " of data visualization
* Created to replicate the plots of Matlab

## Matplotlib has two ways to plot functions
1. Functional programming
2. OOP

```python
#1. Functional programming
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x=np.arange(0,9)
```

```python
x
```

```python
y=2*x
```

```python
y
```

```python
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.xlim(0,6)
plt.ylim(0,6)
plt.title("Y=2X Graph")
plt.savefig("My first plot.png")
```

```python
a=np.arange(0,10)
```

```python
b=a**4
```

```python
#2. OOP Method
#Large axis
fig=plt.figure()
axes1=fig.add_axes([0,0,1,1])
axes1.plot(a,b)

axes1.set_xlim(0,8)
axes1.set_ylim(0,5000)
axes1.set_xlabel("a")
axes1.set_ylabel("b")
axes1.set_title("Power of 4")

#Small axis
axes2=fig.add_axes([0.1,0.1,0.25,0.25])
axes2.plot(a,b)

axes2.set_xlim(1,2)
axes2.set_ylim(0,20)
axes2.set_xlabel("a")
axes2.set_ylabel("b")
axes2.set_title("Zoomed in")
```

```python
fig=plt.figure(figsize=(2,2),dpi=200)
axes=fig.add_axes([0,0,1,1])
axes.plot(a,b)

fig.savefig('new_fig.png',bbox_inches='tight')
```

```python
#Subplots

fig,axes=plt.subplots(nrows=2,ncols=2)
axes[0][0].plot(x,y)
axes[0][1].plot(x,y)
axes[1][0].plot(x,y)
axes[1][1].plot(a,b)

fig.suptitle("Figure title",fontsize=20)
#plt.tight_layout()
fig.subplots_adjust(wspace=0.9,hspace=0.8)
fig.savefig("super figure.png")
```

```python
# Legend and styling
x=np.arange(0,11)
```

```python
fig=plt.figure(dpi=100)
ax=fig.add_axes([0,0,1,1])
ax.plot(x,x,label='x vs x',color='#3f4ca1',lw=3,ls=':',marker='o',ms=10,markerfacecolor='yellow',markeredgecolor='blue',markeredgewidth=2)
ax.plot(x,x**2,label="x vs x**2",color='red',ls='--')
ax.legend(loc=(1.1,0.5))
```
